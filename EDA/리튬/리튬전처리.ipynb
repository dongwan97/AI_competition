## 2. 데이터 살펴보기

광물가격데이터에서 다음 날짜가 휴일임에도 불구하고 존재함(제거)

- 2023.06.25

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#가격 종합 환율 시장 희소

tmp=pd.merge(가격2,종합,on='기준일',how='left')
tmp[tmp['광물종합지수'].isna()==True]['기준일']
```

```python
결측일자=['2021-09-18','2021-09-26','2022-04-02','2022-04-24','2022-05-07',
     '2023-04-23','2023-05-06','2023-06-25','2024-02-04','2024-04-07','2024-04-28','2024-05-11']

가격3=가격2[~가격2['기준일'].isin(결측일자)]
가격3
```

```python
가격종합=pd.merge(가격3,종합,on='기준일',how='left')
가격종합.isna().sum()
```

```python
가격종합환율=pd.merge(가격종합,환율,left_on='기준일',right_on='Date',how='left')
가격종합환율.isna().sum()
```

```python
가격종합환율희소=pd.merge(가격종합환율,희소,on='기준일',how='left')
가격종합환율희소.isna().sum()
```

```python
가격종합환율희소=가격종합환율희소[['기준일','기준가격','광물종합지수','희유금속지수','Close']]
가격종합환율희소
```

```python
# 기존 데이터에서 에서 연도와월만 저장하여 시장과 merge

가격종합환율희소['기준연월']=가격종합환율희소['기준일'].str[:7]
가격종합환율희소
```

```python
final=pd.merge(가격종합환율희소,시장,left_on='기준연월',right_on='기준일',how='left')
final.isna().sum()
```

```python
final
```

```python
수산화리튬
```

```python
final2=pd.merge(final,수산화리튬,left_on='기준연월',right_on='기간',how='left')
final2.isna().sum()
```

```python
final3=pd.merge(final2,탄산리튬,left_on='기준연월',right_on='기간',how='left')
final3.isna().sum()
```

```python
final3
```

```python
final4=pd.merge(final3,수급안정화,left_on='기준연월',right_on='기준일',how='left')
final4
```

```python
final5=final4[['기준일_x','기준가격','광물종합지수','희유금속지수','Close','시장전망지표','수산화리튬 수출 중량','수산화리튬 수입 중량',
            '탄산리튬 수출 중량','탄산리튬 수입 중량','무역수지_x','무역수지_y','수급안정화지수']]
final5.rename(columns={'기준일_x':'기준일','Close':'환율','무역수지_x':'수산화리튬_무역수지','무역수지_y':'탄산리튬_무역수지'},inplace=True)
final5
```

```python
final5.to_csv('../data/final/리튬data.csv',encoding='cp949')
```
