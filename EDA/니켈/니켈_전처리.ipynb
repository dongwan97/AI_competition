```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#가격 종합 환율 시장 희소

tmp=pd.merge(가격2,종합,on='기준일',how='left')
tmp.isna().sum()
```

```python
결측일자=['2021-09-18','2021-09-26','2022-04-02','2022-04-24','2022-05-07',
     '2023-04-23','2023-05-06','2023-06-25','2024-02-04','2024-04-07','2024-04-28','2024-05-11']

가격3=가격2[~가격2['기준일'].isin(결측일자)]
가격3
```

```python
가격종합=pd.merge(가격3,종합,on='기준일',how='left')
가격종합.isna().sum()
```

```python
가격종합환율=pd.merge(가격종합,환율,left_on='기준일',right_on='Date',how='left')
가격종합환율.isna().sum()
```

```python
가격종합환율메이저=pd.merge(가격종합환율,메이저,on='기준일',how='left')
가격종합환율메이저.isna().sum()
```

```python
가격종합환율메이저=가격종합환율메이저[['기준일','기준가격','광물종합지수','메이저금속지수','Close']]
가격종합환율메이저
```

```python
# 기존 데이터에서 에서 연도와월만 저장하여 시장과 merge

가격종합환율메이저['기준연월']=가격종합환율메이저['기준일'].str[:7]
가격종합환율메이저
```

```python
final=pd.merge(가격종합환율메이저,시장,left_on='기준연월',right_on='기준일',how='left')
final.isna().sum()
```

```python
수출입.head()
```

```python
수출입.drop(0,axis=0,inplace=True)
수출입.head()
```

```python
final2=pd.merge(final,수출입,left_on='기준연월',right_on='기간',how='left')
final2.head()
```

```python
final2.isna().sum()
```

```python
final3=pd.merge(final2,수급안정화,left_on='기준연월',right_on='기준일',how='left')
final3
```

```python
final4=final3[['기준일_x','기준가격','광물종합지수','메이저금속지수','Close','시장전망지표','무역수지','수출 중량','수출 금액','수입 중량','수입 금액','수급안정화지수']]
```

```python
final4.rename(columns={'기준일_x':'기준일','Close':'환율'},inplace=True)
final4
```

```python
final4.isna().sum()
```
